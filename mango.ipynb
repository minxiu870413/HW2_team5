```python
import os
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from keras.models import Sequential, load_model
from keras.datasets import cifar10
from keras.utils import np_utils, plot_model
from keras.layers import Dense,Dropout,Flatten,Conv2D,MaxPooling2D
```

```python
from os import listdir
mypath = '/content/drive/MyDrive/Train_Image'
files = listdir(mypath)
files.sort()
files
```

```python
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
import numpy as np 
train_image = [] 
for i in range(len(files)): 
  train_img = image.load_img('/content/drive/MyDrive/Train_Image/'+files[i], target_size=(28,28,3), color_mode="rgb")
  train_img = image.img_to_array(train_img)  
  train_img = train_img/255
  train_image.append(train_img) 
X = np.array(train_image)
```

```python
from os import listdir
mypath = '/content/drive/MyDrive/Test_Image'
files = listdir(mypath)
files.sort()
files
```

```python
from keras.preprocessing import image
test_image = [] 
for i in range(len(files)): 
  img = image.load_img('/content/drive/MyDrive/Test_Image/'+files[i], target_size=(28,28,3), color_mode="rgb")
  img = image.img_to_array(img) 
  img = img/255
  test_image.append(img) 
test_image = np.array(test_image)
```

```python
train_data = pd.read_csv(r"train.csv")
test_data = pd.read_csv(r"Test.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
train_data['label'] = labelencoder.fit_transform(train_data['label'])
test_data['label'] = labelencoder.fit_transform(test_data['label'])
```

```python
from keras.utils import to_categorical
y_train = train_data['label'].values 
y_train = to_categorical(y_train)
y_test = test_data['label'].values 
y_test = to_categorical(y_test)
```

```python
model = Sequential() 
model.add(Conv2D(128,kernel_size=(3,3),activation='relu',input_shape=(28,28,3)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.5))
model.add(Conv2D(128,kernel_size=(3,3),activation='relu',input_shape=(28,28,3)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.5))
model.add(Conv2D(128,kernel_size=(3,3),activation='relu',input_shape=(28,28,3)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.5))
model.add(Flatten())
model.add(Dense(units=128, kernel_initializer='normal', activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=3, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
history = model.fit(X, y_train, validation_split=0.1, epochs=30, batch_size=30, verbose=2)
```

```python
model = Sequential() 
model.add(Conv2D(128,kernel_size=(3,3),activation='relu',input_shape=(28,28,3)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Conv2D(128,kernel_size=(3,3),activation='relu',input_shape=(28,28,3)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.5))
model.add(Conv2D(128,kernel_size=(3,3),activation='relu',input_shape=(28,28,3)))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.5))
model.add(Flatten())
model.add(Dense(units=128, kernel_initializer='normal', activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=3, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
history = model.fit(X, y_train, validation_split=0.1, epochs=30, batch_size=10, verbose=2)
```

```python
score2=model.evaluate(test_image,y_test)
print(score2)
```

```python
from sklearn.metrics import classification_report
y_pred=model.predict(test_image)
y_pred=(y_pred>0.5)
print(classification_report(y_test,y_pred))
```
